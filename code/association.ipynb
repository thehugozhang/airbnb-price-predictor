```python
import scipy as spy
from scipy import stats as spystats
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from apyori import apriori
```

```python
df = pd.read_csv("Airbnb-CLEANED.csv", usecols = ['id', 'NAME','house_rules'])
```

```python
stop_words = [ 'stop', 'the', 'by', 'to', 'and', 'a', 'in', 'it', 'is', 'i', 'that', 'had', 'on', 'for', 'were', 'was', 'of', '', ' ', '1.', '2.', '3.', '4.']
```

```python
names = df["NAME"]
names_list = []

for (name) in names:
    house_name_attribute = str(name).lower().split(" ")
    
    for (attribute) in house_name_attribute:
        if (attribute in stop_words):
            house_name_attribute.remove(attribute)
            
    if (len(house_name_attribute) > 1):
        names_list.append(house_name_attribute)
```

```python
name_results = apriori(names_list,
                   min_support=0.0025,
                   min_confidence=0.8,
                   min_lift=0.0,
                   max_length=None)

for name_result in name_results:
    print("\n")
    print(name_result)
```

```python
rules = df["house_rules"]
rules_list = []

for (rule) in rules:
    house_rules_attribute = str(rule).lower().split(" ")
    
    for (rule) in house_rules_attribute:
        if (rule in stop_words):
            house_rules_attribute.remove(rule)
            
    if (len(house_rules_attribute) > 1):
        rules_list.append(house_rules_attribute)
        
```

```python
rule_results = apriori(rules_list,
                   min_support=0.05,
                   min_confidence=0.9,
                   min_lift=0.0,
                   max_length=None)

for rule_result in rule_results:
    print("\n")
    print(rule_result)
```
